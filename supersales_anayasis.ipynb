```python
def null_summary(data):
    s = data.isna().sum()
    pct = (s / len(data) * 100).round(2)
    return pd.DataFrame({"missing_count": s, "missing_pct": pct}).sort_values("missing_count", ascending=False)
```

```python
data['Item_Fat_Content'] = data['Item_Fat_Content'].replace({
    'LF': 'Low Fat',
    'low fat': 'Low Fat',
    'reg': 'Regular'
}).astype(str)
```

```python
data.loc[data['Item_Visibility'] == 0, 'Item_Visibility'] = np.nan
```

```python
display(null_summary(data))
display(data[['Item_Fat_Content','Item_Visibility']].head(8))
```

```python
wt_map = data.groupby('Item_Type')['Item_Weight'].mean()
data['Item_Weight'] = data['Item_Weight'].fillna(data['Item_Type'].map(wt_map))
```

```python
data['Item_Weight'] = data['Item_Weight'].fillna(data['Item_Weight'].mean())
display(null_summary(data))
```

```python
vis_map = data.groupby('Item_Identifier')['Item_Visibility'].mean()
data['Item_Visibility'] = data['Item_Visibility'].fillna(data['Item_Identifier'].map(vis_map))
data['Item_Visibility'] = data['Item_Visibility'].fillna(data['Item_Visibility'].mean())

display(null_summary(data))
```

```python
CURRENT_YEAR = 2025
data['Outlet_Age'] = CURRENT_YEAR - data['Outlet_Establishment_Year'].astype(int)
```

```python
data['Price_per_unit_wt'] = data['Item_MRP'] / data['Item_Weight']
display(null_summary(data))
```

```python
data['Visibility_Ratio'] = data['Item_Visibility'] / data['Item_Visibility'].mean()
display(null_summary(data))
```

```python
data['Is_Low_Fat'] = (data['Item_Fat_Content'] == 'Low Fat').astype(int)
#data[['Outlet_Age','Price_per_unit_wt','Visibility_Ratio','Is_Low_Fat']].describe().T
```

```python
# validation 
checks = []

# 1. no negative sales
checks.append(("negative_sales_count", (data['Item_Outlet_Sales'] < 0).sum()))

# 2. reasonable MRP
checks.append(("mrp_min_max", (data['Item_MRP'].min(), data['Item_MRP'].max())))

# 3. no nulls after imputation
checks.append(("nulls_after_impute", data.isna().sum().to_dict()))

for name, result in checks:
    print(f"{name}: {result}")
```

```python
#graph
sales_by_outlet = data.groupby('Outlet_Type')['Item_Outlet_Sales'].sum().sort_values(ascending=False)
display(sales_by_outlet)

# bar chart
plt.figure()
sales_by_outlet.plot(kind='bar')
plt.title("Total Sales by Outlet Type")
plt.ylabel("Total Sales")
plt.xlabel("Outlet Type")
plt.ticklabel_format(style='plain', axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.scatter(data['Item_MRP'], data['Item_Outlet_Sales'], alpha=0.3)
plt.title("MRP vs Sales (scatter)")
plt.xlabel("Item MRP")
plt.ylabel("Item Outlet Sales")
plt.tight_layout()
plt.show()

# binned summary
data['MRP_bin'] = pd.qcut(data['Item_MRP'], 10, duplicates='drop')
binned = data.groupby('MRP_bin')['Item_Outlet_Sales'].mean().reset_index()
display(binned)
```

```python

top_items = data.groupby('Item_Type')['Item_Outlet_Sales'].sum().sort_values(ascending=False).head(15)
display(top_items)

plt.figure()
top_items.plot(kind='bar')
plt.title("Top 15 Item Types by Sales")
plt.ylabel("Total Sales")
plt.xticks(rotation=45, ha='right')
#plt.tight_layout()
plt.show()
```

```python
# cell 14
# Save transformed data for Power BI
out_path = "bigmart_transformed_pbi.csv"
data.to_csv(out_path, index=False)
print("Saved transformed CSV:", out_path)

# create a few aggregated CSVs that are useful for dashboards
data.groupby(['Outlet_Type','Outlet_Size'])['Item_Outlet_Sales'].sum().reset_index().to_csv("agg_outlet_sales.csv", index=False)
data.groupby('Item_Type')['Item_Outlet_Sales'].sum().reset_index().to_csv("agg_itemtype_sales.csv", index=False)
print("Saved aggregated CSVs: agg_outlet_sales.csv, agg_itemtype_sales.csv")
```
